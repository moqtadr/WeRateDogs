## Cleaning

```python
#Copy Data files before cleaning
tweet_ar_cleaned = tweet_ar.copy()
```

```python
#Define
#745 name are none â€“ 55 a and we will not need the column 
#Delete unnecessary data columns ( in_reply_to_status_id, in_reply_to_user_id, retweeted_status_id, retweeted_status_user_id, retweeted_status_timestamp )
```

```python
#Code
#Delete unnecessary data columns
tweet_delete_columns = ['in_reply_to_status_id','name' ,'expanded_urls' , 'in_reply_to_user_id', 'retweeted_status_id', 'retweeted_status_user_id', 'retweeted_status_timestamp']
tweet_ar_cleaned.drop(tweet_delete_columns, axis = 1, inplace=True)
```

```python
#Test
tweet_ar_cleaned.info()
```

```python
#Define  Timestamp is in object data type not DateTime format
```

```python
#Code change timestamp column to the correct data type - datetime
tweet_ar_cleaned['timestamp'] = pd.to_datetime(tweet_ar_cleaned['timestamp'])
```

```python
#Define Puppo, Pupper, Floofer and doggo represent the dog type or category ( One Variable ) so we will combine all of them in one column 
```

```python
# Code
# Make 4 columns of dogcategory Puppo, Pupper, floofer and doggo one column
# Delete None values of dog_category column
tweet_ar_cleaned =  tweet_ar_cleaned.replace('None', '')
tweet_ar_cleaned['dog_category'] = tweet_ar_cleaned['doggo']+tweet_ar_cleaned['pupper']+tweet_ar_cleaned['puppo']+tweet_ar_cleaned['floofer']

```

```python
dogs_categ = ['doggo','puppo', 'floofer', 'pupper']
tweet_ar_cleaned.drop(dogs_categ, axis = 1, inplace=True)
```

### DataSet 2 

```python
#Create a copy of datat to clean
image_prediction_cleaned = image_prediction.copy()
```

```python
#Define Three different predictions for the same tweet
```

```python
# Code function to pick the right prediction between the 3
prediction = []
confidence = []

def prediction_confidence(dataframe):
    if dataframe['p1_dog'] == True:
        prediction.append(dataframe['p1'])
        confidence.append(dataframe['p1_conf'])
    elif dataframe['p2_dog'] == True:
        prediction.append(dataframe['p2'])
        confidence.append(dataframe['p2_conf'])
    elif dataframe['p3_dog'] == True:
        prediction.append(dataframe['p3'])
        confidence.append(dataframe['p3_conf'])
    else:
        prediction.append(np.nan)
        confidence.append(np.nan)

image_prediction_cleaned.apply(prediction_confidence, axis=1)
image_prediction_cleaned['prediction'] = prediction
image_prediction_cleaned['confidence'] = confidence
```

```python
#drop the unecessary columns and rows without predictions
image_prediction_cleaned = image_prediction_cleaned.drop(['img_num', 'p1', 'p1_conf', 'p1_dog', 'p2', 'p2_conf', 'p2_dog', 'p3', 'p3_conf', 'p3_dog'], 1)
image_prediction_cleaned.dropna(subset = ['prediction', 'confidence'], inplace = True)
```
